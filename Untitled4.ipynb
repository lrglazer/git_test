```python
import pandas as pd # data tables
import numpy as np # math
import matplotlib.pyplot as plt # visualizations
import seaborn as sns # visualizations
```

```python
# lodad the dataset from a CSV file into a pandas DataFrame
df = pd.read_csv("collegiate_athlete_injury_dataset.csv")
# Display entire DataFrame
df
```

```python
# Display the first 10 rows of the DataFrame to get a preview of the data
df.head(10)
```

```python
# Display a random sample of 10 rows from the DataFrame
df.sample(10)
```

```python
#Display a concise summary of DataFrame, including data types and non-null values
df.info()
```

```python
# Generate Descriptive Statistics
df.describe()
```

```python
# Clean up data...check for missing values in each column and sum them up (check for)
df.isnull().sum()
```

```python
# Remove rows with empty cells
df = df.dropna()
# iterates through each text-based column to print its name and the frequency of each unique item within it.
for col in df.columns:
    if df[col].dtype == 'object':
        print(f"Column: {col}")
        print(df[col].value_counts())
        print("\n")
```

```python
#Count the occurreces of each unique value in the Injury_Indicator column and plot it as a bar chart
df["Injury_Indicator"].value_counts().plot(kind = "bar")
# Set the title of the plot
plt.title("Injury Distribution")
# Display the plot
plt.show()
```

```python
# Create a box plot to visualize the distribution of Fatigue_Score for each Injury_Indicator
sns.boxplot(x = "Injury_Indicator", y = "Fatigue_Score", data=df)
# Set the title of the plot
plt.title("Fatigue vs Injury")
# Display the plot
plt.show()
```

```python
# Fatigue score is higher for injured athletes
```

```python
# Select only numeric columns for correlatin analysis
num_cols = [
    "Age",
    "Height_cm",
    "Weight_kg", # Fixed typo 'Wieght_kg' to 'Weight_kg'
    "Training_Intensity",
    "Training_Hours_Per_Week",
    "Recovery_Days_Per_Week",
    "Match_Count_Per_Week",
    "Rest_Between_Events_Days", # Added missing comma
    "Fatigue_Score",
    "Performance_Score",
    "Team_Contribution_Score",
    "Load_Balance_Score",
    "ACL_Risk_Score", # Fixed typo 'ACL_Risk_score' to 'ACL_Risk_Score'
    "Injury_Indicator"]

# Filter num_cols to ensure only existing columns are included
num_cols = [c for c in num_cols if c in df.columns]
# Calculate the correlation matrix...
corr = df[num_cols].corr()
plt.figure(figsize=(12,9)) # Corrected figsize to be a tuple
sns.heatmap(corr, annot=True, cmap = 'coolwarm', center = 0)
plt.title('Correlation Heatmap (Numeric Features)') # Corrected typo 'Numperic' to 'Numeric'
plt.show()
```
